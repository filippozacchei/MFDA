```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np

def plot_trace(trace, var_names=None, figsize=(12, 6), save_path=None):
    """
    Plots the trace of MCMC samples for given variables.

    Parameters:
    - trace: ArviZ InferenceData object containing MCMC samples.
    - var_names: List of variable names to plot. If None, all variables are plotted.
    - figsize: Tuple specifying the figure size.
    - save_path: If provided, saves the plot to this path.
    """
    az.plot_trace(trace, var_names=var_names, figsize=figsize)
    plt.tight_layout()  
    if save_path:
        plt.savefig(save_path)      
    plt.show()

file_path = 'samples_1-step_1.nc'  # Replace with your file path
trace = az.from_netcdf(file_path)
plot_trace(trace, var_names=None, figsize=(12, 6), save_path='trace_plot.png')
```
